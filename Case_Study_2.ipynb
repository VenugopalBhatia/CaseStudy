Q2: Top 5 Investors by number of investments

```python
from heapq import nlargest
data.InvestorsName.fillna('N/A',inplace = True)
Investors = data.InvestorsName.str.strip()
Investors = data.InvestorsName.str.split(',')


#print(Investors)
InvDict = {}
#Investors = Investors.fillna('U')
#Investors = Investors.astype(str)
for i in Investors:
    for j in i:
        j = j.strip()
        InvDict[j] = InvDict.get(j,0) + 1
print(InvDict['Undisclosed'])
TopInvestors = nlargest(5,InvDict,key = InvDict.get)
```

```python
print(TopInvestors)
li = []
for i in TopInvestors:
    print(i,InvDict[i])
    li.append(InvDict[i])
```

```python
explode = (0.1,0.15,0.2,0.25,0.3)
plt.pie(li,explode = explode,labels = TopInvestors,shadow = True,radius = 1.5,autopct = '%.2f',colors = ('r','#164B93','g','#DF5810','#931621'))
plt.axis('equal')
plt.show()
```
